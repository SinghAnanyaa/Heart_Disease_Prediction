# Heart Disease Prediction Using Machine learning

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing and Reading the dataset

```python
df = pd.read_csv('framingham.csv')
df.head(12) 
```

## Data Analysis

```python
df.shape #gives the no. of rows and columns
```

```python
df.describe() #to find mean,deviation,25% quartile etc.
```

```python
df.isna().sum() #prints null values for each rows
```

### Removing NaN / NULL vales from the data

```python
df.dropna(axis = 0, inplace = True) 
print(df.shape) #removes rows wilth null values in any column
```

```python
df['TenYearCHD'].value_counts() 
```

## Data Visualization

### Correlation Matrix

```python
plt.figure(figsize = (14, 10)) 
sns.heatmap(df.corr(), cmap='Oranges',annot=True, linecolor='gray', linewidths=1.0)
plt.show()
```

```python
sns.catplot(data=df, kind='count', x='male',hue='currentSmoker')
plt.show() #Countplot of people based on their sex and whether they are Current Smoker or not
```

### Countplot - subplots of No. of people affecting with CHD on basis of their sex and current smoking.

```python
sns.catplot(data=df, kind='count', x='TenYearCHD', col='male',row='currentSmoker', palette='Blues')
plt.show() 
```

### Separating the data into feature and target data.

```python
X = df.iloc[:,:-1]  #not including the last column 'TenYearCHD'
y = df.iloc[:,-1] #taking only 'TenYearCHD' as it is our target variable
```

```python
X.head()
```

```python
y.head()
```

### Importing the model and assigning the data for training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=21) #splitting dataset
```

### Applying the ML model - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=2000)
logreg.fit(X_train, y_train) ##‘fit’ method trains the algorithm on the training data, after the model is initialized
```

```python
#Predicting the test set result
y_pred_lr = logreg.predict(X_test) #predicts the target value(y) for the 30% testing data which means we predict y value for the 
#testing data
#trhough the training data
```

### Predicting the accuracy score

```python
#score = logreg.score(y_test, y_pred)
#print("Accuracy score is:",score) 
# Display accuracy score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score
sc_lr=accuracy_score(y_test, y_pred_lr) #compare original y value of testing data and predicted value
print('Accuracy score is: ',sc_lr)
```

## Getting the Confusion Matrix and Classification Report

### Classification Report

```python
print("\nPrecision Score is: " , precision_score(y_test, y_pred_lr)) #printing the precision score
print("Recall Score is: " , recall_score(y_test, y_pred_lr)) #printing the recall score
print("Classification Report is:\n\n",classification_report(y_test,y_pred_lr))
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred_lr) 
print("Confusion Matrix is:\n",cm)
conf_matrix = pd.DataFrame(data = cm,  
                           columns = ['Predicted:0', 'Predicted:1'],  
                           index =['Actual:0', 'Actual:1']) 
plt.figure(figsize = (10, 6)) 
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens", linecolor="Blue", linewidths=1.5) 
plt.show() 
```

### Applying the ML model - RandomForest

```python
from sklearn.ensemble import RandomForestClassifier  
classifier= RandomForestClassifier(n_estimators= 10, criterion="entropy")  
classifier.fit(X_train, y_train) 

#n_estimators= The required number of trees in the Random Forest. The default value is 10. We can choose any number but need to take care of the overfitting issue.
#criterion= It is a function to analyze the accuracy of the split. Here we have taken "entropy" for the information gain.
```

```python
#Predicting the test set result  
y_pred_rf= classifier.predict(X_test)  
```

```python
# Display accuracy score
sc_rf=accuracy_score(y_test, y_pred_rf)
print('Accuracy score is: ',sc_rf)
```

## Getting the Confusion Matrix and Classification Report

### Classification Report

```python
print("\nPrecision Score is: " , precision_score(y_test, y_pred_rf)) #printing the precision score
print("Recall Score is: " , recall_score(y_test, y_pred_rf)) #printing the recall score
print("Classification Report is:\n\n",classification_report(y_test,y_pred_rf))
#Precision quantifies the number of positive class predictions that actually belong to the positive class.
#Recall quantifies the number of positive class predictions made out of all positive examples in the dataset.
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred_rf) 
print("Confusion Matrix is:\n",cm)
conf_matrix = pd.DataFrame(data = cm,  
                           columns = ['Predicted:0', 'Predicted:1'],  
                           index =['Actual:0', 'Actual:1']) 
plt.figure(figsize = (10, 6)) 
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens", linecolor="Blue", linewidths=1.5) 
plt.show() 
```

### Applying the ML model - K Nearest Neighbour

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

inertias = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X_train)
    inertias.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k') #we can use the elbow method to find the optimal value of k at which accuracy is highest
plt.show()
#To determine the optimal number of clusters, we have to select the value of k at the “elbow” 
#ie the point after which the distortion/inertia start decreasing in a linear fashion
```

```python
# K Nearest Neighbours Classifier

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5) #create an object of k nearest neighbours classifier

knn.fit(X_train, y_train) #training model using this fit method
y_pred_knn = knn.predict(X_test) #store the predicted values on X_test
```

```python
#display accuracy score
sc_knn = accuracy_score(y_test , y_pred_knn)
print('Accuracy score is: ',sc_knn)
```

## Getting the Confusion Matrix and Classification Report

### Classification Report

```python
print("\nPrecision Score is: " , precision_score(y_test, y_pred_knn)) #printing the precision score
print("Recall Score is: " , recall_score(y_test, y_pred_knn)) #printing the recall score
print("Classification Report is:\n\n",classification_report(y_test,y_pred_knn))
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred_knn) 
print("Confusion Matrix is:\n",cm)
conf_matrix = pd.DataFrame(data = cm,  
                           columns = ['Predicted:0', 'Predicted:1'],  
                           index =['Actual:0', 'Actual:1']) 
plt.figure(figsize = (10, 6)) 
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens", linecolor="Blue", linewidths=1.5) 
plt.show() 
```

### Applying the ML model - Decision Tree Classifier

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier #create an object of decision tree classifier

dt = DecisionTreeClassifier(criterion = 'gini' , max_depth = 3) #3levels in tree for training model

dt = dt.fit(X_train , y_train) #training model using this fit method
y_pred_dt = dt.predict(X_test) #store the predicted values on X_test
```

```python
#display accuracy score
sc_dt = accuracy_score(y_test , y_pred_dt)
print('Accuracy score is: ',sc_dt)
```

## Getting the Confusion Matrix and Classification Report

### Classification Report

```python
print("\nPrecision Score is: " , precision_score(y_test, y_pred_dt)) #printing the precision score
print("Recall Score is: " , recall_score(y_test, y_pred_dt)) #printing the recall score
print("Classification Report is:\n\n",classification_report(y_test,y_pred_dt))
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred_dt) 
print("Confusion Matrix is:\n",cm)
conf_matrix = pd.DataFrame(data = cm,  
                           columns = ['Predicted:0', 'Predicted:1'],  
                           index =['Actual:0', 'Actual:1']) 
plt.figure(figsize = (10, 6)) 
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens", linecolor="Blue", linewidths=1.5) 
plt.show() 
```
